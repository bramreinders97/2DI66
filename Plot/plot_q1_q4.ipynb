```python
import matplotlib.pyplot as plt

def plot_confidence_interval(x, prob, ci, color='#2187bb', horizontal_line_width=0.25):

    left = x - horizontal_line_width / 2
    right = x + horizontal_line_width / 2
    plt.plot([x,x], [ci[0], ci[1]], color=color)
    plt.plot([left, right], [ci[0], ci[0]], color=color)
    plt.plot([left, right], [ci[1], ci[1]], color=color)
    plt.plot(x, prob, '.', color='#f44336')
```

```python
def create_plots_with_draw(white_prob, white_ci, black_prob, black_ci, draw_prob, draw_ci, Q):
    plt.xticks([1, 2, 3], ['White', 'Black', 'Draw'])
    plt.title(f"Win probabilities Q{Q}")
    plot_confidence_interval(1, white_prob, white_ci)
    plot_confidence_interval(2, black_prob, black_ci)
    plot_confidence_interval(3, draw_prob, draw_ci)
    plt.show()

def create_plots_no_draw(white_prob, white_ci, black_prob, black_ci, Q):
    plt.xticks([1, 2], ['White', 'Black'])
    plt.title(f"Win probabilities Q{Q}")
    plot_confidence_interval(1, white_prob, white_ci)
    plot_confidence_interval(2, black_prob, black_ci)
    plt.show()
```
